```python
data1 = pd.read_csv("2022-05-27-15-38-27-fsr1.csv", delimiter=',')
data2 = pd.read_csv("2022-05-27-15-38-27-fsr2.csv", delimiter=',')
data2.head()
```

```python
time = data1["time"]
fsr1 = data1["\data"]
fsr2 = data2[".data"]
```

```python
figure(figsize=(12, 6), dpi=80)

plt.plot(time, fsr1, label="FSR_Kanan")
plt.plot(time, fsr2, label="FSR_Kiri")

plt.xlabel('Time')
plt.ylabel('FSR Data')
plt.title("Pembacaan data sensor FSR")

plt.xlim([60, 270])
plt.legend()

plt.show
```

```python
time = vel1['time']
vel2 = vel2['.x']
vel1 = vel1['.x']
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(time, vel2, label = 'Velocity with feedback')
ax.plot(time, vel1, label = "Velocity no feedback")
ax.plot(time, [0.025] * 76 , label = "Setpoint")

ax.set_xlabel('Time')
ax.set_ylabel('Velocity')
ax.set_title('Velocity Per Step')

ax.set_xlim([52, 58])
ax.set_ylim([0.00, 0.05])

ax.yaxis.set_major_locator(MultipleLocator(0.05))
ax.yaxis.set_minor_locator(MultipleLocator(0.025))
ax.grid(b=True, which='minor')

ax.legend()
```
